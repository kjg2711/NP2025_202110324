print() 함수
- 숫자나 문자열 등의 데이터나 수식을 화면에 출력

```python
print(170 + 65)
# 문자열 출력
print("김종규")
# 문자열 3번 출력
print("김종규" * 3)
# 변수 a, b 값 출력
a = 170 
b = 65
print(a, b)
# 문자열과 변수의 조합 
print('키', a, '몸무게', b)
```

print() 함수 서식 지정
- 숫자나 문자열 등의 데이터를 서식을 지정하여 화면에 출력

서식 
- %d : 10진수 정수로 출력
- %f : 실수로 출력
- %s : 문자열 출력
- %c : 한 글자 출력
- %x : 16진수 정수로 출력
- %0 : 8진수 정수로 출력

```python
num = 25
print("num = % d" %num)
h = float(input("h: "))
print("h=%.1f"%h)
```

관계 연산자(relation operator)
- 비교연산자 
- 항의 대소관계 또는 동등 관계 판정
- 연산 결과는 True 또는 False

    + x < y
        + x가 y보다 작으면 True
    + x <= y 
        + x가 y보다 작거나 같으면 True
    + x > y 
        + x가 y보다 크면 True
    + x >= y
        + x가 y보다 크거나 같으면 True
    + x == y
        + x와 y가 같으면 True
    + x != y
        + x와 y가 다르면 True

    

논리 연산자(logical operator)
- 논리곱 (and) , 논리합 (or) , 논리 부정 (not)
- 논리 연산의 대상은 참 또는 거짓. 결과도 참 또는 거짓 

    - x and y
        - x 와 y가 모두 참일때만 연산결과가 참 
    - x or y
        - x 와 y 둘중 하나라도 참이면 연산 결과가 참 
    - not x 
        - x 의 부정. x가 참(거짓) 이면 결과는 거짓(참). 

멤버 연산자(membership operator)
- in
- 특정 문자열 또는 값이 문자열 또는 리스트(반복객체)에 속해 있는지 판별하는 연산자
    - 'py' in 'python'
    - 'ty' not in 'python'

식별 연산자(identity operator)
- is : 양쪽 변수의 주소가 같은지 판단
    - if id(a) == id(b)
- is not : 양쪽 변수의 주소가 다른지 판단
    - if id(a) != id(b)
