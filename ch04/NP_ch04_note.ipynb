```python
## 변수 초기화 부분 ##
outStr = "" 
count, i = 0, 0

inStr = input("Type string : ")
count = len(inStr) 

for i in range(0, count) :
    outStr += inStr[count - (i + 1)] 
print("Reversed string : %s" % outStr)
```

### 집합(set)
- 빈 집합을 만들 때는 set() 함수를 이용
- set에 항목의 개수를 셀 때는 len() 함수 
- 항목을 추가할 때는 add() 함수
- 제거할 때는 removeO 함수

- 집합의 생성 
    




```python
# 1부터 45까지의 수 중에서 6개를 선택하여 로또 번호를 만드는 프로그램
import random
pick = set() 
while len(pick) < 6:
 n = random.randint(1,45) 
 if n not in pick:
    pick.add(n) 
print(pick)
print(sorted(pick)) 
```
